```python
import os
import cv2
import numpy as np
from random import shuffle
from tqdm import tqdm

IMG_SIZE = 50  # 假設的圖像大小
TRAIN_DIR = r'C:\Users\Aran\Desktop\貓狗\train'  # 訓練資料目錄
TEST_DIR = r'C:\Users\Aran\Desktop\貓狗\test'  # 測試資料目錄

def label_img(img):
    word_label = img.split('.')[0]  # 根據 . 進行分割，取第一部分
    if word_label == 'cat': 
        return [1, 0]
    elif word_label == 'dog': 
        return [0, 1]
    else:
        raise ValueError(f"Unknown label: {word_label}")

def create_train_data():
    training_data = []
    for img in tqdm(os.listdir(TRAIN_DIR)):
        try:
            label = label_img(img)
            path = os.path.join(TRAIN_DIR, img)
            img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            if img_data is None:
                print(f"Error loading image: {path}")
                continue
            img_data = cv2.resize(img_data, (IMG_SIZE, IMG_SIZE))
            training_data.append([np.array(img_data), np.array(label)])
        except Exception as e:
            print(f"Error processing file {img}: {e}")
    shuffle(training_data)
    np.save('train_data.npy', training_data)
    return training_data

def create_test_data():
    testing_data = []
    for img in tqdm(os.listdir(TEST_DIR)):
        try:
            path = os.path.join(TEST_DIR, img)
            img_num = img.split('.')[0]
            img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            if img_data is None:
                print(f"Error loading image: {path}")
                continue
            img_data = cv2.resize(img_data, (IMG_SIZE, IMG_SIZE))
            testing_data.append([np.array(img_data), img_num])
        except Exception as e:
            print(f"Error processing file {img}: {e}")
    shuffle(testing_data)
    np.save('test_data.npy', testing_data)
    return testing_data

train_data = create_train_data()
test_data = create_test_data()
```
